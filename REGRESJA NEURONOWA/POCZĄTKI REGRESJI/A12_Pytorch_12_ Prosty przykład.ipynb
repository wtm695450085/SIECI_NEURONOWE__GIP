```python
import torch
import torch.nn as nn
```

```python
# Defining input size, hidden layer size, output size and batch size respectively
n_in, n_h, n_out, batch_size = 5, 5, 1, 5
```

```python
# Create dummy input and target tensors (data)
x = torch.tensor([[-1.2,  0.9,  0.7, -1.1,  0.1],[ 0.5,  0.2,  0.5, -1.3,  0.6],[-0.9, -0.8,  0.5, -0.5,  0.7],[-0.4,  1.3, -1.7, -0.7, -2.1],[-0.3,  0.9,  2.7,  0.6, -1.4]], dtype=torch.float)
y = torch.tensor([[1.0], [0.0], [0.0], [1.0], [1.0]])
```

```python
print('x:',x.shape)
print('y:',y.shape)
```

<span style="font-size: 130%;color:#1155cc"> Model sieci neuronwej

```python
# Create a model
model = nn.Sequential(nn.Linear(n_in, n_h),
                     nn.ReLU(),
                     nn.Linear(n_h, n_out),
                     nn.Sigmoid())
```

<span style="font-size: 130%;color:#1155cc"> Funkcja straty i optymalizer

```python
# Construct the loss function
loss_fn = torch.nn.MSELoss()

# optimizer z krokiem uczącym lr = 0.01 (Stochastic Gradient Descent in this case)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)


# można użyć innego optymalizera
#learning_rate = 0.1
#optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# Gradient Descent
for epoch in range(1500):
    y_pred = model(x)

    # Compute and print loss
    loss = loss_fn(y_pred, y)
    print('epoch: ', epoch,' loss: ', loss.item())

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    
    # perform a backward pass (backpropagation)
    loss.backward()
    
    # Update the parameters
    optimizer.step()
```
